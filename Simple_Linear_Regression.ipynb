# **Aims to predict continuous real numbers like salary, temperature.**

Simplest ML Model

Ordinary Least Squares

yi = actual amount

y^i = wouldve been amount

**Import Library**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Dataset**

```python
dataset = pd.read_csv('/content/Salary_Data.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
# .values is a numpy array
# print((x))
# makes it 2d
```

**Split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=0)

#test size = percentage of data for x_test and y_test
#random state = takes random rows
```

**Training SLR on training set**

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
#regression = predict continuous real value
regressor.fit(x_train, y_train)
# fitting finds m and c (y = mx + c)
# m = coefficient
# m and c are training parameters

```

```python
print(regressor.intercept_) # C
print(regressor.coef_) # M
```

**Predicting results**

```python
y_pred = regressor.predict(x_train)
#"predict" method always expects a 2D array as the format of its inputs.
#12 -> Scalar, [12] 1D, [[12]] 2D
```

**Visualize training **

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, y_pred , color = 'blue')
plt.title('Salary vs Experience')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.show()
```

```python
y_pred = regressor.predict(x_test)
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_test, y_pred, color = 'blue')
plt.title('Salary vs Experience')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.show()
```

```python
y_pred = regressor.predict(x_train)
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, y_pred, color = 'blue')
plt.title('Salary vs Experience')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.show()
```
